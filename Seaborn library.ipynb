```python
#importing required libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
Iris = sns.load_dataset("iris")

#construct iris plot
ax = sns.swarmplot(x='species', y='petal_length', data=Iris)
plt.show()
```

```python
#using pairplot to show relation between features
sns.pairplot(iris)
plt.show()
```

```python
#plot tips dataset using violin plot
tips = sns.load_dataset('tips')
sns.violinplot(x='total_bill', data=tips)
plt.show()
```

```python
import numpy as np
from scipy import stats, integrate
sns.set(color_codes=True)

np.random.seed(sum(map(ord,"distibutions")))
x = np.random.normal(size=100)

#by default, this will draw a histogram and fit a kernal density estimate
sns.distplot(x);

#fitting parametric distribution
sns.distplot(x, kde=False, fit=stats.gamma)

plt.show()
```

```python
import pandas as pd

sns.set(style='whitegrid', palette='muted')
iris = sns.load_dataset('iris')

#"Melt(one or more columns are identifier variables, while all other columns, considered measured variables)"-
#convert the dataset to long form or tidy representation
iris = pd.melt(iris, "species", var_name="measurement")
print(iris.head())

#draw categorical scatter plot to show each observation
sns.swarmplot(x='measurement', y='value', hue='species', data=iris)
plt.show()
```

```python
#draw linear regression model using tips dataset
sns.set(color_codes=True)
tips = sns.load_dataset('tips')
print(tips.head())
sns.lmplot(x='total_bill', y='tip', hue='smoker',data=tips)
plt.show()
```

```python
#boxplot
sns.set(style='ticks')
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='PRGn')
sns.despine(offset=10, trim=False)
plt.show()
```
